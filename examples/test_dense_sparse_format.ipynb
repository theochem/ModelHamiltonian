Initializing dense array

```python
norb = 2
nelec = norb
hubbard = Hubbard([(f"C{i}",f"C{i+1}", 1) for i in range(1,norb)], alpha=0, beta=-1, 
                  u_onsite=np.array([1 for i in range(norb)]))
ecore, h, v = hubbard.get_hamilton()
```

```python
print('printing one-electron integrals as a dense array:\n', h)
print('printing two-electron integrals as a dense array:\n', v)
```

```python
print('converting one-elctron integrals in dense array to sparse array format and printing:')
hs = hubbard.to_sparse(h); print(hs)
print('converting two-elctron integrals in dense array to sparse array format and printing:')
vs = hubbard.to_sparse(v); print(vs)
```

```python
print('converting one-elctron integrals in sparse array to dense array format and printing:')
hd = hubbard.to_dense(hs); print(hd)
print('converting two-elctron integrals in sparse array to dense array format and printing:')
vd = hubbard.to_dense(vs); print(vd)
print('converting two-elctron integrals in sparse array to dense array format and printing:')
vd = hubbard.to_dense(vs,4); print(vd)

print(hd == h)
print(vd == v)
```
